```python
import os
import numpy as np
```

```python
LOAD_DDF_BIT    = 1 << 0
EQU_DDF_BIT     = 1 << 1
COLLIDE_BIT     = 1 << 2
BOUNCE_BACK_BIT = 1 << 3
STORE_DDF_BIT   = 1 << 4

DUMP_RHO_BIT    = 1 << 27
DUMP_VX_BIT     = 1 << 28
DUMP_VY_BIT     = 1 << 29
DUMP_VZ_BIT     = 1 << 30
```

```python
def leftInletRightOutlet(nx:int=32, ny:int=32, nz:int=32, leftVel:float=0.05, folder="../data/left_inlet_right_outlet"):
    folder += f"_{nx}_{ny}_{nz}_init_state"
    if not os.path.exists(folder):
        os.makedirs(folder)
    flags = np.zeros((nz, ny, nx), dtype=np.uint32)
    flags[1:-1,1:-1,1:-1] = (LOAD_DDF_BIT|COLLIDE_BIT|STORE_DDF_BIT|DUMP_RHO_BIT|DUMP_VX_BIT|DUMP_VY_BIT|DUMP_VZ_BIT)
    flags[:, 0,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[:,-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[0,1:-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[-1,1:-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[1:-1,1:-1,0] = (EQU_DDF_BIT|STORE_DDF_BIT)
    flags[1:-1,1:-1,-1] = (EQU_DDF_BIT|STORE_DDF_BIT)
    flags.tofile(folder+"/flag.dat")

    vx = np.zeros((nz, ny, nx), dtype=np.float32)
    vx[1:-1,1:-1,0] = leftVel                                                #left surface
    vx.tofile(folder+"/vx.dat")
```

```python
def leftInletRightOutletCubeObs(cx_min:int,cy_min:int,cz_min:int,cx_max:int,cy_max:int,cz_max:int,nx:int=288,ny:int=272,nz:int=280,left_vel:float=0.05,folder_prefix="left_inlet_right_outlet_cube_obs")->None:
    folder = folder_prefix + f"_{nx}_{ny}_{nz}_init_state"
    if not os.path.exists(folder):
        os.makedirs(folder)
    flags = np.zeros((nz, ny, nx), dtype=np.uint32)
    flags[1:-1,1:-1,1:-1] = (LOAD_DDF_BIT|COLLIDE_BIT|STORE_DDF_BIT|DUMP_RHO_BIT|DUMP_VX_BIT|DUMP_VY_BIT|DUMP_VZ_BIT)
    flags[:, 0,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[:,-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[0,1:-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[-1,1:-1,:] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags[1:-1,1:-1,0] = (EQU_DDF_BIT|STORE_DDF_BIT)
    flags[1:-1,1:-1,-1] = (EQU_DDF_BIT|STORE_DDF_BIT)
    flags[cz_min:cz_max,cy_min:cy_max,cx_min:cx_max] = (LOAD_DDF_BIT|BOUNCE_BACK_BIT|STORE_DDF_BIT)
    flags.tofile(folder+"/flag.dat")

    vx = np.zeros((nz, ny, nx), dtype=np.float32)
    vx[1:-1,1:-1,0] = left_vel                                            #left surface
    vx.tofile(folder+"/vx.dat")
```

```python
nx,ny,nz = 288,272,280
cx_min = nx//2-nx//6
cy_min = ny//2-ny//6
cz_min = nz//2-nz//6
cx_max = nx//2+nx//6
cy_max = ny//2+ny//6
cz_max = nz//2+nz//6
leftInletRightOutletCubeObs(cx_min,cy_min,cz_min,cx_max,cy_max,cz_max)
```
