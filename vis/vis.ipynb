```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
NUM_NODE = 2
NUM_STEP = 4
DLT_STEP = 15
SHAPE = (4,8,32)
data = dict()
for nodeIdx in range(NUM_NODE):
    for step in range(NUM_STEP):
        arr = np.fromfile(f"../data/vx_{step*DLT_STEP:05d}_{nodeIdx}.dat", dtype=np.float32).reshape(*SHAPE)
        data[(step,nodeIdx)] = arr
```

```python
data[(3,1)][:,4,:]
```

```python
def showArr2d(data, title):
    plt.imshow(data)
    plt.colorbar()
    plt.title(title)
    plt.show()
```
